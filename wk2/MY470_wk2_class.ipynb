# Strings

* Ordered sequences of characters
* Immutable #means that if I call some methods to change, I need to save it to a variable, otherwise nothing change. 

```python
x = 'my string'

# Capitalises the first character (chr) of the string
x = x.capitalize()

# prints the string
print(x)

# prints the chr at index 3
print(x[3])

# prints the last chr
print(x[-1])

# print a range 
# NOTE: not inclusive of the last index, 4 chrs because python starts at 0
print(x[0:4])

# Index one to the last index
# Again, not inclusive
print(x[1:-1])

# EXTENDED SLICING 
# Get every other (2) item in the string.
print(x[::2])

# Reverse steps, every other chr
print(x[::-2])
```

```python
# Exercise 3: Remove the trailing white space in the string below, 
# replace all double spaces with single space, and format to a sentence 
# with proper punctuation. Print the resulting string.

string1 = '  this  is a very badly.  formatted string -  I would  like to make it cleaner\n'
s2 = string1.strip().replace('  ',' ').replace('badly.','badly').capitalize().replace('i would','I would') + '.'
print(s2)

#from the teacher
#capitalize come firstly
s3 = string1.strip().capitalize().replace(' i ',' I ').replace('  ',' ')
s3 = s3.replace('.','') + '.'
print(s3)
```

```python
# Exercise 4: Convert the string below to a list

s = "['apple', 'orange', 'pear', 'cherry']"
#s2 = s.replace("['","'").replace("']","'")
#print(s2)
#from the teacher
#use the lstrip and rstrip to remove a character, which are more efficient than replace
ls = s.replace("'",'').lstrip('[').rstrip(']').split(', ')
type(ls)
```

```python
# Exercise 5: Reverse the strings below.

s1 = 'stressed'
s2 = 'drawer'
#string doesnot have a reverse function;list has reverse/reversed function
s3 = s1[::-1]
s4 = s2[::-1]
print(s3,s4)

#another sol
s5 = ''.join(reversed(s1))
#reversed(s1) return a list but can not see,list(reversed(s1)) then we can see the list
print(s5)
```
